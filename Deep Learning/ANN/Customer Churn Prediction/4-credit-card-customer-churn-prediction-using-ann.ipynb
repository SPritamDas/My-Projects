## Introduction

In today’s competitive financial ecosystem, customer retention has become a vital metric for sustaining profitability. Credit card churn, where customers stop using a company’s services, leads to significant revenue loss. Predicting and mitigating this churn is a top priority for financial institutions, especially since acquiring new customers is more expensive than retaining current ones.

Leveraging **machine learning** to predict customer churn allows businesses to stay ahead of the competition. **Artificial Neural Networks (ANNs)**, with their ability to learn complex patterns from large datasets, are one of the most powerful tools available for churn prediction.

> **Why Predict Customer Churn?**
> - Minimize customer acquisition costs
> - Improve marketing and retention strategies
> - Increase customer lifetime value


---

## Problem Statement

The objective of this project is to build a robust ANN model capable of predicting whether a credit card customer is likely to churn. The prediction is based on a wide array of customer data, including demographics, account activity, transaction history, and interactions with customer service.

By applying **deep learning techniques**, this project aims to:
- **Accurately predict churn probability**
- **Identify key drivers of churn**
- **Reduce false positives and false negatives**

*Typical Credit Card Churn Prediction Workflow:*
1. Data collection and preprocessing
2. Feature engineering
3. ANN model development
4. Model evaluation and tuning
5. Prediction and insights generation

---

## Why Choose ANN for Churn Prediction?

### 🔥 **Advantages of ANN:**
- **Non-linear Modeling**: Captures complex relationships in customer behavior that traditional models like logistic regression might miss.
- **Feature Learning**: Automatically extracts relevant patterns from raw input without manual feature selection.
- **Scalability**: Can handle large datasets, making it ideal for credit card customer data.
- **High Dimensionality**: Deals well with both categorical and numerical data.

Traditional machine learning models, such as decision trees or support vector machines, may struggle with high-dimensional, non-linear data. In contrast, ANNs excel in identifying subtle trends and correlations within the data, making them a superior choice for churn prediction.

---

## Data Overview

The dataset used for this project includes the following features:

- **Demographics**: Age, gender, income level, etc.
- **Account Information**: Account tenure, credit limit, balance, and payment history.
- **Transaction Patterns**: Frequency of transactions, transaction amounts, types of purchases.
- **Customer Service**: Number of support calls, complaints, and response times.

Each feature plays a vital role in predicting the likelihood of churn, and careful preprocessing and normalization are required to feed these inputs into the ANN model.

---

## Model Architecture

The **ANN architecture** for this project consists of:

- **Input Layer**: Feeds customer data (demographics, transactions, etc.) into the network.
- **Hidden Layers**: Multiple fully connected layers, where the model learns intricate patterns.
- **Output Layer**: A binary classification node, predicting whether the customer will churn or not.

*Training Procedure*:
1. **Backpropagation**: Used to adjust weights by minimizing the error (loss function).
2. **Activation Function**: ReLU in hidden layers, Sigmoid in the output layer.
3. **Optimization**: Gradient Descent is employed to optimize model parameters.

## Balancing the Dataset using SMOTE

```python
import imblearn
from imblearn import over_sampling
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=42)
x_res, y_res = sm.fit_resample(x, y)
```

## Creating Training and Test Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_res,y_res,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

# Creating Model

## Importing Libraries

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model_1 = Sequential()
```

##### Model Description
* 1-Hidden Layer
* In Hidden Layer 3 Perceptrons

# Model - 1

```python
## Hidden Layer-1(No.of Perceptron,Activation_Function, How many input will be given to each perceptron = No.of features in X)
model_1.add(Dense(3,activation='relu',input_dim = 11))
## Output Layer
model_1.add(Dense(1,activation='sigmoid'))
```

```python
model_1.summary()
```

It means 
1. Output shape is in the given layer how many output we will get = No. of nodes/perceptrons in each layer.
2. Parameters = No. of trainable parameters = Weights + Biases

```python
model_1.compile(loss='binary_crossentropy',optimizer = 'Adam')
```

```python
model_1.fit(x_train_scaled,y_train,epochs=10)
```

### Trainable Parameters of 1st Layer (Hidden)

```python
model_1.layers[0].get_weights()
```

#### Trainable Prameters of Output Layers

```python
model_1.layers[1].get_weights()
```

## Predictions

```python
y_log = model_1.predict(x_test_scaled)
```

```python
y_log
```

Here the output does not belong to {0,1} but belonds to [0,1].
Because, we have used `Sigmoid` as activation Finction. 
To covert these outputes to `0` or `1`, we need to use a `cut-off`.

```python
from sklearn.metrics import RocCurveDisplay,roc_auc_score
cutoff = [i/100 for i in range(100)]
roc_auc = np.array([])
for i in range(len(cutoff)):
    y_pred= np.where(y_log> cutoff[i],1,0)
    roc_auc=np.append(roc_auc,roc_auc_score(y_test,y_pred))
plt.figure(figsize=(10,3))
sns.lineplot(y = roc_auc,x = cutoff)
plt.title('AUC Vs Threshold')
plt.show()
AUC_Vs_Threshold = pd.DataFrame([cutoff,roc_auc]).T
AUC_Vs_Threshold.head(3)
print(AUC_Vs_Threshold[AUC_Vs_Threshold.iloc[:,1]==max(AUC_Vs_Threshold.iloc[:,1])])
```

```python
y_pred = np.where(y_log>0.57,1,0)
```

```python
y_pred
```

### Accuracy

```python
from sklearn.metrics import accuracy_score,f1_score
print('Accuracy:',accuracy_score(y_test,y_pred))
print('f1_Score:',f1_score(y_test,y_pred))
```

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred)).plot()
plt.title('Confusion Matrix for Model-1')
plt.show()
```
